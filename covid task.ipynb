# import libraries

```python

import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline

```

# Data Reading

```python

df = pd.read_excel(r"C:\Users\rehna\Desktop\covid task\file.xlsx")
df


```

# removing some columns

```python
x=df.drop(["Death","LOS_Y","Age","Black","Severity","White","Ferritin","Ferritin > 300","CrctProtein","Procalcitonin","Procalciton > 0.1","Troponin","Troponin > 0.1"],axis=1)


x
```

```python
y=df["Death"]
y
```

# spliting training and testing

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.25, random_state = 0)


```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
xtrain = sc_x.fit_transform(xtrain)
xtest = sc_x.transform(xtest)

print (xtrain[0:10, :])
```

# using logicsticRegression

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(xtrain, ytrain)
```

```python
y_pred = classifier.predict(xtest)
```

# confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest, y_pred)

print ("Confusion Matrix : \n", cm)
```

# Accuracy

```python
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(ytest, y_pred))
```
